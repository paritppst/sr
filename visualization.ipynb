```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('./data.csv')
data['Revenue'] = data['Qsold'] * data['P_price']
data.head()
```

```python
sns.set_style('whitegrid')
sns.set_context('paper')
cp = sns.countplot(x='S_type', data=data, palette='Blues')
cp.set_title('Number of Official Shop VS User', size=15)
cp.set_xlabel('Shop Type',fontsize=13.5)
cp.set_ylabel('Count', fontsize=13.5)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

```python
sns.set_style('whitegrid')
box = sns.catplot(x='S_type', y = 'Revenue', data=data, kind='box', palette='Blues')
box.fig.suptitle('Revenue of Official Mall VS User', y = 1.03, fontsize=15)

plt.xlabel('Shop Type', fontsize=13.5)
plt.ylabel('Revenue', fontsize=13.5)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

```python
sns.set_context('notebook')
sns.set_style('white')
plt.figure(figsize=(10,6))
scatter = sns.scatterplot(x='Qsold', y='Revenue', data=data, palette= 'plasma' ,hue='S_type', hue_order=['Official Mall', 'User'], alpha=0.5)
scatter.set_title('Relationship Between Revenue and Quantity sold', y =1.025, fontsize=15)
scatter.set_xlabel('Qsold', fontsize = 13.5)
scatter.set_ylabel('Revenue', fontsize=13.5)
plt.setp(scatter.get_legend().get_texts(), fontsize='11')
plt.setp(scatter.get_legend().get_title(), fontsize='15')
plt.show()
```

```python
data_om = data.loc[data['S_type']=="Official Mall"]
data_user = data.loc[data['S_type']=="User"]
```

```python
sns.set_context('notebook')
sns.set_style('whitegrid')
g = sns.catplot(x='P_score', data=data, kind='count', col='S_type', palette="Blues")
g.fig.suptitle("Product Score Official Mall VS User", fontsize=14.5, y=1.05)
g.set_titles("{col_name}", fontsize = 14.5)
plt.locator_params(axis='y', nbins=10)
plt.locator_params(axis='x', nbins=14)
g.axes[0,0].set_xlabel('Product Score', fontsize=12)
g.axes[0,1].set_xlabel('Product Score', fontsize=12)
g.axes[0,0].set_ylabel('Count', fontsize =12)
plt.show()
```

```python
sns.set_context('paper')
sns.set_style('white')
scatter = sns.lineplot(x='P_score', y='S_follower', data=data, palette= 'plasma' ,hue='S_type', hue_order=['Official Mall', 'User'], alpha=0.5)
scatter.set_title('Relationship Between Revenue and Quantity sold', y =1.025, fontsize=15)
scatter.set_xlabel('Qsold', fontsize = 13.5)
scatter.set_ylabel('Revenue', fontsize=13.5)
plt.show()
```

```python
sns.set_context('notebook')
sns.set_style('whitegrid')
plt.hist(x='EPP', data=data_user, bins=10)
plt.hist(x='EPP', data=data_om,bins=2)
plt.legend(['User', 'Official Mall'])
plt.title('EPP of Official Mall VS User', fontsize=15, y=1.05)
plt.xlabel('EPP',fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
